```python
top_batsmen = df.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
top_batsmen.plot(kind='bar', figsize=(10,6), color='orange')
plt.title("Top 10 Run Scorers in IPL")
plt.ylabel("Total Runs")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df1['result_margin'], bins=30, kde=True)
plt.title("Distribution of Matches Won by Runs")
plt.show()
```

```python
toss_win_match_win = df1['toss_winner'] == df1['winner']
sns.countplot(x=toss_win_match_win)
plt.title("Impact of Toss on Match Winning")
plt.show()
```

```python
df.groupby(['batter', 'non_striker'])['total_runs'].sum().sort_values(ascending=False).head(10).plot(kind='bar', figsize=(10,6))
plt.title("Top Batting Partnerships in IPL")
plt.show()
```

```python
team_wins=df1["winner"].value_counts()
total_matches=df1["team1"].value_counts() + df1["team2"].value_counts()
win_percentage=team_wins/total_matches*100
win_percentage
```

```python
win_percentage.plot(kind="pie",figsize=(10,6))
plt.title("win percentage")
plt.show()
```

```python
df1.groupby('season')['result_margin'].sum().plot(kind='line', marker='o', figsize=(10,6))
plt.title("Total Runs Scored Over IPL Seasons")
plt.ylabel("Runs")
plt.xlabel("Season")
plt.show()
```

```python
df['total_balls'] = df.groupby('bowler')['ball'].transform('count')
bowler_stats = df.groupby('bowler')['total_runs'].sum() / (df.groupby('bowler')['total_balls'].sum() / 6)
bowler_stats = bowler_stats.sort_values().head(10) 

bowler_stats.plot(kind='barh', figsize=(10,6), color='purple')
plt.title("Top 10 Bowlers (Best Economy Rate)")
plt.xlabel("Economy Rate")
plt.show()
```
